```python
import urllib.request
import json
import re
from collections import Counter
```

```python
def find_words(text):
    return re.findall('[а-яА-Я]+', text.lower())
```

```python
ids = ['1', '168747430', '173450']
lst = []
for i in ids:
    req = urllib.request.Request('https://api.vk.com/method/wall.get?owner_id=' + i + '&count=10&v=5.73&access_token=b45e92d1b45e92d1b45e92d1bcb43fcc9abb45eb45e92d1eed3d845c49175940031ee9f')
    response = urllib.request.urlopen(req)
    result = response.read().decode('utf-8')
    vk_resp = json.loads(result)
    for j in range(10):
        lst.extend(find_words(vk_resp['response']['items'][j]['text']))
dct = list(Counter(lst).most_common())
dct
```

```python
dct = dct[:20]
dct
```

```python
import matplotlib.pyplot as plt
```

```python
X, Y = [], []
xlables = []
for i in range(len(dct)):
    X.append(i)
    Y.append(dct[i][1])
    xlables.append(dct[i][0])

#X = [1,   1.5, 2,   3, 5, 6]
#Y = [1.8, 1,   3.6, 2, 1, 3]

plt.plot(X, Y)
plt.show()
```

```python
plt.scatter(X, Y, color='green')
#plt.bar(X, Y, color='green')
plt.title("Частотный словарь постов ВК")
plt.xlabel("Частотность слова", color='green')
plt.ylabel("Рейтинг слова", color='green')
plt.xticks(X, xlables, rotation=60)
plt.show()
```
